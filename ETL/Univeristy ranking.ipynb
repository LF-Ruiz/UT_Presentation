```python
url2 = 'http://archive.shanghairanking.com/ARWU2020.html'
rank2 = pd.read_html(url2)
```

```python
df1 = rank2[0]
df1
```

```python
df1=df1[['Institution*','World Rank','National/Regional Rank']]
df1
```

```python
df3 = pd.read_csv('resources/txcolleges.csv')
df3
```

```python
tx = df3.merge(df1, left_on = 'instname', right_on = 'Institution*')
tx
```

```python
tx = tx.dropna(axis='columns')
tx
```

```python
tx.info()
```

```python
tx2 = tx[['Institution*', 'city', 'zip', 'World Rank', 'National/Regional Rank']].sort_values('World Rank', ignore_index = True)
tx2
```

```python
tx2 = tx2.rename(columns = {'Institution*' : 'institution', 'World Rank' : 'wrank', 'National/Regional Rank': 'nrank'})
tx2
```

```python
tx2.info()
```

```python
wr = tx2['wrank']
nr = tx2['nrank']
```

```python
a = wr.str.split("-", expand=True)
b = nr.str.split("-", expand=True)
```

```python
tx2['wrank2'] = a[0]
tx2['nrank2'] = b[0]
# ['B', 'C']
tx2.drop(columns=['wrank', 'nrank'], axis=1)
tx2
```

```python
tx2['wrank'] = tx2['wrank2'].astype('int32')
tx2['nrank'] = tx2['nrank2'].astype('int32')
tx2 = tx2.drop(['wrank2', 'nrank2'], axis=1)
tx2
```

```python
tx2.to_csv('resources/txcollegesranking.csv', index = False)
```
